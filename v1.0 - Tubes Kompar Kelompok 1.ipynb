```python
from google.colab import files
import pandas as pd
import io
import matplotlib.pyplot as plt
```

```python
uploaded = files.upload()
file_name = next(iter(uploaded))
```

```python
df = pd.read_csv(io.BytesIO(uploaded[file_name]))
```

```python
print(df.head())
```

```python
print(df.info())
```

```python
total_revenue_per_category = df.groupby('Product Category')['Revenue'].sum().sort_values(ascending=True)
```

```python
total_revenue_per_category.plot(kind='bar', xlabel='Product Category', ylabel='Total Revenue')

plt.title('Total Revenue per Category')
plt.show()
```

```python
total_revenue_per_subcategory = df.groupby('Sub Category')['Revenue'].sum().sort_values(ascending=True)
```

```python
total_revenue_per_subcategory.plot(kind='bar', xlabel='Sub Category', ylabel='Total Revenue')

plt.title('Total Revenue per Sub Category')
plt.show()
```

```python
total_revenue_per_country = df.groupby('Country')['Revenue'].sum().sort_values(ascending=True)
```

```python
total_revenue_per_country.plot(kind='bar', xlabel='Country', ylabel='Total Revenue')

plt.title('Total Revenue per Country')
plt.show()
```

```python
total_revenue_per_age = df.groupby('Customer Age')['Revenue'].sum().sort_values(ascending=True)
```

```python
total_revenue_per_age.tail().plot(kind='bar', xlabel='Customer Age', ylabel='Total Revenue')

plt.title('Total Revenue per Customer Age')
plt.show()
```

```python
total_revenue_per_month = df.groupby('Month')['Revenue'].sum().sort_values(ascending=True)
```

```python
total_revenue_per_month.plot(kind='bar', xlabel='Month', ylabel='Total Revenue')

plt.title('Total Revenue per Month')
plt.show()
```

```python
total_revenue_per_gender = df.groupby('Customer Gender')['Revenue'].sum().sort_values(ascending=True)
```

```python
total_revenue_per_gender.plot(kind='pie', xlabel='Customer Gender', ylabel='Total Revenue')

plt.title('Total Revenue per Customer Age')
plt.show()
```
